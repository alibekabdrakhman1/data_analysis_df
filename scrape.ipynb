```python
import pandas as pd   
import requests       
from bs4 import BeautifulSoup 
import numpy as np 
import re
import random

book_names = []
prices = []
year = []
ratings = []
categories = []
url = ""
for i in range(1, 5):
    if i == 1 :
        url = "https://www.flip.kz/catalog?subsection=2&filter-show=1"
    else:
        url = "https://www.flip.kz/catalog?subsection=2&filter-show=1&page=" + str(i)
    print(url) 
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    book_data = soup.find_all('div', class_='good-list-item')
#     print(book_data)
    
   
    
    
    for item in book_data:   
        name = item.find("div","info").a.get_text().strip()
        book_names.append(name)
        
        price = item.find('div', 'price').get_text().split('₸')
        prices.append(price[0].strip())
        rating = item.find('div','comment').a.get_text().strip()
        ratings.append(rating)
        
        add_info = item.find('div','add-info').find_all('span')

        year_of_realese = add_info[0].get_text().split(',')
        year.append(year_of_realese[1].strip())
        
        category = add_info[1].get_text().strip()
        categories.append(category)
        
        
        

print(len(book_names))
print(len(prices))
print(len(year))
print(len(categories))

book_DF = pd.DataFrame({'Name of book': book_names, 'Year of relase': year, 'Price': prices, 'Category': categories, 'Rating': ratings})
```
